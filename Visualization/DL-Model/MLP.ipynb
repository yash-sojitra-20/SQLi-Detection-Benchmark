```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# === MLP Results ===
results = [
    {
        'label': 'Original',
        'accuracy': 0.9948253557567918,
        'precision': 0.9955,
        'recall': 0.9934,
        'f1': 0.9944,
        'conf_matrix': np.array([[3903, 5], [27, 2249]])
    },
    {
        'label': 'RUS Balanced',
        'accuracy': 0.9918734900065891,
        'precision': 0.9919,
        'recall': 0.9919,
        'f1': 0.9919,
        'conf_matrix': np.array([[2267, 10], [27, 2249]])
    },
    {
        'label': 'ROS Balanced',
        'accuracy': 0.9955214331413947,
        'precision': 0.9955,
        'recall': 0.9955,
        'f1': 0.9955,
        'conf_matrix': np.array([[3897, 11], [24, 3883]])
    }
]

# Convert to DataFrame
results_df = pd.DataFrame(results)

# === 1. Accuracy Bar Plot ===
plt.figure(figsize=(8, 5))
sns.barplot(x='label', y='accuracy', data=results_df, palette='Purples_r')
plt.title('MLPClassifier - Accuracy Comparison')
plt.ylim(0.98, 1.0)
plt.ylabel('Accuracy')
plt.xlabel('Dataset')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# === 2. Macro Precision, Recall, F1 Bar Plot ===
metrics_df = results_df.melt(id_vars='label', value_vars=['precision', 'recall', 'f1'],
                             var_name='Metric', value_name='Score')

plt.figure(figsize=(10, 6))
sns.barplot(data=metrics_df, x='label', y='Score', hue='Metric', palette='Set1')
plt.title('MLPClassifier - Macro Avg Precision, Recall, F1-score')
plt.ylim(0.98, 1.0)
plt.ylabel('Score')
plt.xlabel('Dataset')
plt.legend(title='Metric')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# === 3. Confusion Matrices ===
for res in results:
    plt.figure(figsize=(5, 4))
    sns.heatmap(res['conf_matrix'], annot=True, fmt='d', cmap='Purples', cbar=False)
    plt.title(f"Confusion Matrix - {res['label']}")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()
```
