```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../SQLi_Dataset.csv")

# Count label distribution
label_counts = df['Label'].value_counts()

# Plot
plt.figure(figsize=(4, 4))
plt.bar(label_counts.index.astype(str), label_counts.values, color=['#4CAF50', '#F44336'])

plt.title("SQLi Dataset Distribution (Benign vs Malicious)", fontsize=14)
plt.xlabel("Class Label", fontsize=12)
plt.ylabel("Number of Samples", fontsize=12)

# Add text labels on top of bars
for i, v in enumerate(label_counts.values):
    plt.text(i, v + 100, str(v), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Hatching patterns for black & white printing
hatches = ['////', '....']

plt.figure(figsize=(4, 4))

bars = plt.bar(label_counts.index.astype(str),
               label_counts.values,
               color="white",              # white fill
               edgecolor="black",          # black outlines
               hatch=hatches[0:len(label_counts)])  # hatching

plt.title("SQLi Dataset Distribution (Benign vs Malicious)", fontsize=14)
plt.xlabel("Class Label", fontsize=12)
plt.ylabel("Number of Samples", fontsize=12)

# Add text labels on top of bars
for i, v in enumerate(label_counts.values):
    plt.text(i, v + 100, str(v), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
